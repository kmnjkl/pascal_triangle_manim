Functions:

```python
def comb(n, k):
    if 0 <= k <= n:
        nn = 1
        kk = 1
        for t in range(1, min(k, n - k) + 1):
            nn *= n
            kk *= t
            n -= 1
        return nn // kk
    else:
        return 0
```

# Pascal's triangle

## Definition

```python
# Triangle
class PascalTriangle:
  def __init__(self, size, frame_height):
    self.size = size
    self.numbers = [[comb(n, m) for m in range(n+1)] for n in range(size)]
    
    self.mo = VGroup(*[VGroup(*[Integer(self.numbers[n][m]) for m in range(n+1)]).arrange(RIGHT, buff=0) for n in range(size)])
    nw = self.mo[0][0].width
    w0=nw
    # Row buff multiplier
    nwm = 6
    n=0
    for i in self.mo:
      if n==0: 
        n += 1
        nw += nwm*w0
        continue
      w = i.width
      buff_row = (nw-w)/n
      i.arrange(RIGHT, buff=buff_row)
      # i.arrange_in_grid(rows=1, cols=n+1, col_width=20)
      nw += nwm*w0
      n+=1
    buff_col = 0.5
    self.mo.arrange(DOWN, buff=buff_col).scale_to_fit_height(frame_height-1)

    #print(self.numbers)
# tri = PascalTriangle(4)
```

# Fibonacci

```python
from numpy import e
# Fibonacci numbers
def fib(num):
  if num <= 0:
    print("fib(): UNEXPECTED NOT POSITIVE PARAMETER")
  elif num == 1 or num == 2:
    return 1
  else:
    return fib(num-1) + fib(num-2)
```

```python
%%manim -ql -v WARNING MyScene

sc_right = 7+1/9

class MyScene(Scene):
  def construct(self):
    self.play(Create(Text(" ")))
    self.wait(2)
    # self.clear()
    self.play(FadeOut(*self.mobjects, *self.foreground_mobjects))

    self.clear()
    pt = PascalTriangle(10, self.camera.frame_height)
    self.play(Create(pt.mo))
    self.wait(1.5)

    self.play(pt.mo.animate.scale_to_fit_height(6).align_on_border(DL))
    self.wait(0.5)

    colors = (BLUE_C, YELLOW_C, MAROON_C, GREEN_C, TEAL_C, GOLD_C, PURPLE_C, RED_C)
    colors_fib = (BLUE_B, YELLOW_B, MAROON_B, GREEN_B, TEAL_B, GOLD_B, PURPLE_B, RED_B)

    fn = 1
    for nm1 in range(pt.size):
      cur_color = colors[nm1%len(colors)]
      fib_sum = VGroup()
      for i in range(nm1//2 + 1):
        pt_cur_num = pt.mo[nm1-i][i]
        self.play(pt_cur_num.animate.set_color(cur_color), run_time=0.2)
        self.play(Indicate(pt_cur_num, scale_factor=1.3), color=cur_color, run_time=0.3)

        if i == nm1//2:
          fib_sum.add(pt_cur_num.copy())
          self.play(fib_sum.animate.arrange(RIGHT).next_to(pt.mo, RIGHT))
          cur_num = Integer(fib(fn), font_size=152).set_color(colors_fib[nm1%len(colors_fib)])
          self.wait(0.5)
          fib_sum.add(Text("=").scale_to_fit_height(pt_cur_num.height), cur_num)
          self.play(fib_sum.animate.arrange(RIGHT).next_to(pt.mo, RIGHT))
          self.wait(0.5)
          cur_num_x_coord = ((8*16/9)/pt.size) * (nm1 + 0.5) + (-7-1/9)
          fib_sum -= cur_num
          self.play(cur_num.animate.scale(0.4).align_on_border(UP).set_x(cur_num_x_coord), FadeOut(fib_sum))
        else:
          fib_sum.add(pt_cur_num.copy(), Text("+").scale_to_fit_height(pt_cur_num.height))
          self.play(fib_sum.animate.arrange(RIGHT).next_to(pt.mo, RIGHT))
          self.wait(0.5)

      fn += 1
    
    self.wait(4)
```
